```python
import pandas as pd
import numpy as np
from transformers import BertModel, BertConfig, PreTrainedTokenizer, BertTokenizer
from datetime import datetime

pd.options.display.expand_frame_repr = True
tok = BertTokenizer.from_pretrained("bert-base-cased")
```

```python
all_features = ["text_tokens", "hashtags", "tweet_id", "present_media", "present_links", "present_domains",\
                "tweet_type","language", "tweet_timestamp", "engaged_with_user_id", "engaged_with_user_follower_count",\
               "engaged_with_user_following_count", "engaged_with_user_is_verified", "engaged_with_user_account_creation",\
               "engaging_user_id", "engaging_user_follower_count", "engaging_user_following_count", "engaging_user_is_verified",\
               "engaging_user_account_creation", "engagee_follows_engager"]
all_labels = ["reply_timestamp", "retweet_timestamp", "retweet_with_comment_timestamp", "like_timestamp"]

all_columns = all_features + all_labels
```

```python
samples = []
with open("part-00000.tsv", encoding="utf-8") as f:
    for line in f.readlines():
        line = line.strip()
        features = line.split("\x01")
        samples.append(features)
```

```python
df = pd.DataFrame(samples,columns=all_columns,copy=False)
df.head(10)
```

```python

print(tok.convert_tokens_to_string(tok.convert_ids_to_tokens(df.iloc[df["text_tokens"].map(len).argmax(),0].split("\t"))))
print(tok.convert_tokens_to_string(tok.convert_ids_to_tokens(df.iloc[123,0].split("\t"))))
```
